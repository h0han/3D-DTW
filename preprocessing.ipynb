```python
def split_string(input_string):
    for i in input_string.split('\n'):
        sp_list.append(i)
        for j in sp_list:
            buf = []
            for k in j.split(' '):
                buf.append(float(k))
                buf_list.append(buf)
    return sp_list, buf_list
```
